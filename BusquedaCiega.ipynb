```python
import time
from tkinter import *
```

```python

class Busquedas():
    def __init__(self,origen,destino,algoritmo):
        self.origen=origen
        self.destino=destino
        self.algoritmo=algoritmo
        self.Tiempo=0
        self.agenda=[] ##Los nodos por explorar
        self.frontera_busqueda=[]##los nodos que ya viste
        self.camino=[]
        self.costoUniforme=False
        self.grafo={'Ambato':[('Riobamba',56,60),('Guaranda',95,120),('Guayaquil',120,240),("Cuenca",100,200),("Milagro",70,140),("Aluasi",30,60),("Azogues",200,400)],
                    'Quevedo':[('Riobamba',90,180),('Guaranda',120,240),('Guayaquil',176,320),("Cuenca",80,160),("Milagro",30,60),("Aluasi",90,180),("Azogues",150,300)],
                    'Guaranda':[('Riobamba',70,140),('Guayaquil',100,200),("Cuenca",80,160),("Milagro",130,230),("Aluasi",50,100),("Azogues",220,440)],
                    'Milagro':[('Riobamba',80,160),('Guaranda',45,90),('Guayaquil',40,80),("Cuenca",144,280),("Aluasi",80,160),("Azogues",140,280)],
                    'Riobamba':[('Guaranda',56,110),('Guayaquil',200,400),("Cuenca",258,500),("Milagro",186,350),("Aluasi",89,180),("Azogues",200,400)],
                    'Aluasi':[('Riobamba',70,140),('Guaranda',90,180),('Guayaquil',220,440),("Cuenca",30,60),("Milagro",170,320),("Azogues",158,300)],
                    'Azogues':[('Riobamba',220,440),('Guaranda',250,500),('Guayaquil',270,520),("Cuenca",250,500),("Milagro",220,440),("Aluasi",30,60)],
                    'Cuenca':[('Riobamba',140,290),('Guaranda',120,240),('Guayaquil',200,400),("Milagro",220,440),("Aluasi",70,140),("Azogues",270,520)],
        }
       
    def f_sucesor(self,nodo):
        a=self.grafo[nodo]
        lista=a.copy()
        for nodo_lista in lista: 
            for nodo_visitado in self.frontera_busqueda:
                if nodo_lista[0]==nodo_visitado[0] and(nodo_lista in lista):
                    lista.remove(nodo_lista)
        if self.costoUniforme:
            #ordenar lista
            lista=sorted(lista,key=lambda nodo: nodo[1])
        return lista
    def f_objetivo(self,nodo):
        if nodo==self.destino:
            return True
        else:
            return False
    def f_primero_profundidad(self):
        self.costoUniforme=False
        self.agenda.append((self.origen,self.origen))
        Encontrado=False
        while not(Encontrado) and (self.agenda):
            nodo=self.agenda.pop()
            if self.f_objetivo(nodo[0]):
                Encontrado=True
                self.frontera_busqueda.append(nodo)
            else:
                self.frontera_busqueda.append(nodo)
                sucesores=self.f_sucesor(nodo[0])
                while sucesores:
                    sucesor=sucesores.pop()
                    self.agenda.append((sucesor[0], nodo[0]))
    def f_camino(self):
        self.camino=[]
        Nodo=self.frontera_busqueda.pop()
        self.camino.insert(0,Nodo[0])
        while (Nodo[1]!=self.origen) :
            for nodo_lista in self.frontera_busqueda:

                if nodo_lista[0]==Nodo[1]:
                    Nodo=nodo_lista
                    break
            self.camino.insert(0,Nodo[0])
        self.camino.insert(0,self.origen)
        return
    def f_costo(self):
        lista=self.camino.copy()
        origen=lista.pop(0)
        costo=0
        minutos=0
        while lista:
            destino=lista.pop(0)
            for nodo in self.grafo[origen]:
                if nodo[0]==destino:
                    costo+=nodo[1]
                    break
            origen=destino
    
        return costo,minutos
    
    def f_primero_anchura(self,ifCosto):
        self.agenda.append((self.origen,self.origen))
        Encontrado=False
        self.costoUniforme=ifCosto
        while not(Encontrado) and (self.agenda):
            nodo=self.agenda.pop(0)
            if self.f_objetivo(nodo[0]):
                Encontrado=True
                self.frontera_busqueda.append(nodo)
            else:
                self.frontera_busqueda.append(nodo)
                sucesores=self.f_sucesor(nodo[0])
                while sucesores:
                    sucesor=sucesores.pop(0)
                    self.agenda.append((sucesor[0], nodo[0]))
    def f_resultadp(self):
        print(self.algoritmo)
        print("Camino: ",self.camino)
        print("Costo: ",self.f_costo())
        #print(self.Tiempo)
        print("Costo camino:", len(self.camino))

        print("Agenda: ",len(self.agenda))
        print("Frontera de busqueda: ",len(self.frontera_busqueda))

    def Cronometro(self,funcion,*argumentos):
        inicio=time.time()
        funcion(*argumentos)
        fin=time.time()
        self.Tiempo=fin -inicio
        return
        
```

```python
MyBusqueda=Busquedas('Cuenca','Guaranda',"Primero en Anchura")
MyBusqueda.Cronometro(MyBusqueda.f_primero_profundidad)
##MyBusqueda.costoUniforme()
MyBusqueda.f_camino()
MyBusqueda.f_resultadp()
```
